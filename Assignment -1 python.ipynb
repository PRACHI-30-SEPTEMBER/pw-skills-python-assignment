#Que.5 Using a while loop,verify if the number A is purely divisible by number B and if so then how
many times it can be divisible.

```python
a = 20
b = 4

count = 0
temp = a

while temp >= b:
      temp -= b
      count += 1

if temp == 0:
    print(a,"is divisible by",b,"and can be divided",count,"times")
else:
    print(a,"is not divisible by",b)
```

```python
20 is dvisible by 4 and can be divided 5 times
```
